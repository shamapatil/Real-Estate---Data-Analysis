```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\shama\\Documents\\cd4.csv")
```

```python
data1 = df

''' Detection '''
# IQR
Q1 = np.percentile(data1['Price_per_sqft'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(data1['Price_per_sqft'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", data1.shape)
 
# Upper bound
upper = np.where(data1['Price_per_sqft'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(data1['Price_per_sqft'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
data1.drop(upper[0], inplace = True)
data1.drop(lower[0], inplace = True)
 
print("New Shape: ", data1.shape)
```
